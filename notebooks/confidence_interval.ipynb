```python
import random
import decimal

print('Random number with seed 42')
random_list = []
random.seed(42)
for i in range(20):
    # Random number with seed 42

    #random_list = decimal.Decimal(random.randrange(1, 20))/2


    random_list.append(float (random.randrange(1, 20)))
    pass
print(random_list)
```

```python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
mean_value , lower, upper = mean_confidence_interval(random_list)
```

```python
variance = sum([((x - mean_value) ** 2) for x in random_list]) / len(random_list)
std = variance ** 0.5
```

```python
print("mean value = ",mean_value)
print("lower confidence = ",lower)
print("upper confidence = ",upper)
print(std)
print(variance)
```

```python
distribution_values = np.random.normal(mean_value, std, 20)
```
