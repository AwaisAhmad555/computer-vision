```python
annotation_json_file_path = '/content/annotations.json'


with open(annotation_json_file_path) as infile:
    json_information = json.load(infile)

images_paths_dictionary = json_dict = {i["img"]: i for i in json_information}
```

```python
# Utility for reading an image and for getting its annotations.
def get_image(name):
    data = images_paths_dictionary[name]
    img_data = plt.imread(data["img"])
    # If the image is RGBA convert it to RGB.
    if img_data.shape[-1] == 4:
        img_data = img_data.astype(np.uint8)
        img_data = Image.fromarray(img_data)
        img_data = np.array(img_data.convert("RGB"))
    data["img_data"] = img_data

    return data
```
