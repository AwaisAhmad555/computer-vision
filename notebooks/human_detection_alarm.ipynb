```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
results = model('/content/pedestrians walking.jpg')
results_person = model.predict(source="/content/pedestrians walking.jpg", save=True, classes=[0])
```

```python
from PIL import Image

# r = results[0]
r = results_person[0]

im_array = r.plot()  # plot a BGR numpy array of predictions
im = Image.fromarray(im_array[..., ::-1])  # RGB PIL image
display(im)
```

```python
print([int(cls) for cls in r.boxes.cls.cpu().tolist()])

print()
```
