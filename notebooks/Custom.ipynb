```python
t2_list = sorted(glob.glob('/content/drive/MyDrive/Brats17TrainingData/HGG/*/*t2.nii.gz'))
t1ce_list = sorted(glob.glob('/content/drive/MyDrive/Brats17TrainingData/HGG/*/*t1ce.nii.gz'))
flair_list = sorted(glob.glob('/content/drive/MyDrive/Brats17TrainingData/HGG/*/*flair.nii.gz'))
mask_list = sorted(glob.glob('/content/drive/MyDrive/Brats17TrainingData/HGG/*/*seg.nii.gz'))

################################################################################################


dataset_path = "/content/drive/MyDrive/Brats17TrainingData/HGG"
slices_folder_path = "/content/drive/MyDrive/Brats17TrainingData/slices_folder"


if os.path.isdir(slices_folder_path):
    pass
else:
    os.makedirs(slices_folder_path)

print()

print(os.path.isdir(slices_folder_path))

print()


slices_image_folder_paths = []

folders_list = sorted(os.listdir(dataset_path))

for image_folder in folders_list:

  image_folder_path = os.path.join(slices_folder_path,image_folder)


  if os.path.isdir(image_folder_path):
    pass
  else:
    os.makedirs(image_folder_path)
    pass


  slices_image_folder_paths.append(image_folder_path)

  #print(image_folder_path)
  #print()

  pass



print()

slices_image_folder_paths = sorted(slices_image_folder_paths)

print(slices_image_folder_paths)

print()


#######saving slices of each MRI volume in its corresponding image folder ==> (slices_image_folder_paths)


####### t2 modality channel

for idx in range(len(t2_list)-205):


  print(t2_list[idx])
  temp_image_t2=nib.load(t2_list[idx]).get_fdata()

  print()
  print("#################################")
  print(temp_image_t2.shape)
  print()


  t2_folder_path = slices_image_folder_paths[idx] + "/t2"

  print(t2_folder_path)

  if os.path.isdir(t2_folder_path):
    pass
  else:
    os.makedirs(t2_folder_path)
    pass

  for image_slice in range(len(temp_image_t2[0,0,:])):

    single_slice = temp_image_t2[:,:,image_slice]

    #image_slice_path = os.path.join(slices_image_folder_paths[idx],"slice_no_"+str(image_slice))


    np.save(t2_folder_path+'/slice_no_'+str(image_slice)+'.npy',single_slice)

    #print(single_slice.shape)

    pass

  pass

print()
print("end t2 images \n")

######################

####### t1ce modality channel

for idx in range(len(t1ce_list)-205):


  print(t1ce_list[idx])
  temp_image_t1ce=nib.load(t1ce_list[idx]).get_fdata()

  print()
  print("#################################")
  print(temp_image_t1ce.shape)
  print()


  t1ce_folder_path = slices_image_folder_paths[idx] + "/t1ce"

  print(t1ce_folder_path)

  if os.path.isdir(t1ce_folder_path):
    pass
  else:
    os.makedirs(t1ce_folder_path)
    pass

  for image_slice in range(len(temp_image_t1ce[0,0,:])):

    single_slice = temp_image_t1ce[:,:,image_slice]

    #image_slice_path = os.path.join(slices_image_folder_paths[idx],"slice_no_"+str(image_slice))


    np.save(t1ce_folder_path+'/slice_no_'+str(image_slice)+'.npy',single_slice)

    #print(single_slice.shape)

    pass

  pass


print()
print("end t1ce images \n")

######################

####### flair modality channel

for idx in range(len(flair_list)-205):


  print(flair_list[idx])
  temp_image_flair=nib.load(flair_list[idx]).get_fdata()

  print()
  print("#################################")
  print(temp_image_flair.shape)
  print()


  flair_folder_path = slices_image_folder_paths[idx] + "/flair"

  print(flair_folder_path)

  if os.path.isdir(flair_folder_path):
    pass
  else:
    os.makedirs(flair_folder_path)
    pass

  for image_slice in range(len(temp_image_flair[0,0,:])):

    single_slice = temp_image_flair[:,:,image_slice]

    #image_slice_path = os.path.join(slices_image_folder_paths[idx],"slice_no_"+str(image_slice))


    np.save(flair_folder_path+'/slice_no_'+str(image_slice)+'.npy',single_slice)

    #print(single_slice.shape)

    pass

  pass

print()
print("end flair images \n")

######################

####### mask/ground_truth

for idx in range(len(mask_list)-205):


  print(mask_list[idx])
  temp_image_mask=nib.load(mask_list[idx]).get_fdata()

  print()
  print("#################################")
  print(temp_image_mask.shape)
  print()


  mask_folder_path = slices_image_folder_paths[idx] + "/ground_truth"

  print(mask_folder_path)

  if os.path.isdir(mask_folder_path):
    pass
  else:
    os.makedirs(mask_folder_path)
    pass

  for image_slice in range(len(temp_image_mask[0,0,:])):

    single_slice = temp_image_mask[:,:,image_slice]

    #image_slice_path = os.path.join(slices_image_folder_paths[idx],"slice_no_"+str(image_slice))


    np.save(mask_folder_path+'/slice_no_'+str(image_slice)+'.npy',single_slice)

    #print(single_slice.shape)

    pass

  pass


```
