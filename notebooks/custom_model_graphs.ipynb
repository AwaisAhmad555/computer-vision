```python
import pandas as pd
csv_file_path = "/content/drive/MyDrive/history.csv"

training_history_dataframe = pd.read_csv(csv_file_path)


##################################################################


#plot the training and validation IoU and loss at each epoch
loss = np.array(training_history_dataframe['loss'])
val_loss = np.array(training_history_dataframe['val_loss'])
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


acc = np.array(training_history_dataframe['iou_Score'])
val_acc = np.array(training_history_dataframe['iou_score_val'])

plt.plot(epochs, acc, 'y', label='Training IOU score')
plt.plot(epochs, val_acc, 'r', label='Validation IOU Score')
plt.title('Training and validation IOU')
plt.xlabel('Epochs')
plt.ylabel('IOU')
plt.legend()
plt.show()

iou = np.array(training_history_dataframe.iloc[:,0])
val_iou = np.array(training_history_dataframe.iloc[:,1])

plt.plot(epochs, iou, 'y', label='Training Accuracy')
plt.plot(epochs, val_iou, 'r', label='Validation Accuracy')
plt.title('Training and validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


print("Epoches : ",len(epochs))
#################################################
```
